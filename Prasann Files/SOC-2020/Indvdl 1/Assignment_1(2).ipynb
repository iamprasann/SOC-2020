```
from random import randint 
from matplotlib import pyplot as plt
import numpy as np

iter=[]
x=[]
for t in range(100): 
  
  arr=[0]*10000
  ones=[1]
  n=0

  arr[5000]=1

  while(True):
    for i in range(10000):
     if(arr[i]==1):
        if(randint(0,19)<7):
          arr[abs(i-1)]=1
        if(randint(0,19)<7):
          arr[(i+1)%10000]=1

    set=np.random.randint(0, 9999, 10)
    for i in range(5):
      arr[set[i]]=arr[set[i]]+arr[set[i+5]]
      arr[set[i+5]]=arr[set[i]]-arr[set[i+5]]
      arr[set[i]]=arr[set[i]]-arr[set[i+5]]

    n=n+1

    ones.append(0)
    
    for i in range(10000):
      if(arr[i]==1):
       ones[n]=ones[n]+1

    if(ones[n]==10000):
      break

  iter.append(n)
  x.append(t+1)

plt.plot(x, iter)
plt.xlabel("#run")
plt.ylabel("#iterations")
plt.show()
plt.savefig("SoC-Plot2-Prasann.jpg")

sum=0
for t in range(100):
  sum=sum+iter[t]

print("avg ", sum/100)
```
