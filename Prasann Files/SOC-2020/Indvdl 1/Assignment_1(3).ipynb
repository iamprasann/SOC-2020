```
from random import randint 
from matplotlib import pyplot as plt
import numpy as np
from scipy.signal import savgol_filter

x=[]
y=[]
for t in range(1): 
  
  arr=[0]*10000
  ones=[1]
  n=0

  arr[5000]=1

  while(True):
    for i in range(10000):
     if(arr[i]==1):
        if(randint(0,19)<7):
          arr[abs(i-1)]=1
        if(randint(0,19)<7):
          arr[(i+1)%10000]=1

    set=np.random.randint(0, 9999, 10)
    for i in range(5):
      arr[set[i]]=arr[set[i]]+arr[set[i+5]]
      arr[set[i+5]]=arr[set[i]]-arr[set[i+5]]
      arr[set[i]]=arr[set[i]]-arr[set[i+5]]

    n=n+1

    ones.append(0)
    
    for i in range(10000):
     ones[n] += arr[i]

    x.append(n)
    y.append(ones[n]-ones[n-1])
     
    if(ones[n]==10000):
      break

plt.plot(x, y)
plt.xlabel("#iteration")
plt.ylabel("dy/dx")
yhat = savgol_filter(y, 51, 2)
plt.plot(x, yhat, color = 'red')
plt.savefig("SoC-Plot3-Prasann.jpg")


```
