```
from random import randint 
from matplotlib import pyplot as plt
import numpy as np
from scipy.signal import savgol_filter

x=[]
y=[]
for t in range(1): 
  
  arr=np.zeros([120,125], dtype=int)
  ones=[1]
  n=0

  arr[59, 62]=1

  while(True):
    for i in range(120):
      for j in range(125):
        if(arr[i, j]==1):
          for p in range(3):
            for q in range(3):
              if(randint(0, 3)<1):
                arr[abs(i-1+p)%120, abs(j-1+q)%125]
          for r in range(5):
            if(randint(0,24)<2):
              arr[abs(i-2), abs(j-2+r)%125]=1
            if(randint(0,24)<2):
              arr[(i+2)%120, abs(j-2+r)%125]=1
          for r in range(3):
            if(randint(0, 24)<2):
              arr[abs(i-1+r)%120, abs(j-2)]=1
            if(randint(0, 24)<2):
              arr[abs(i-1+r)%120, (j+2)%125]=1

    seti=np.random.randint(0, 119, 16)
    setj=np.random.randint(0, 124, 16)
    for i in range(8):
      arr[seti[i], setj[i]]=arr[seti[i], setj[i]]+arr[seti[i+8], setj[i+8]]
      arr[seti[i+8], setj[i+8]]=arr[seti[i], setj[i]]-arr[seti[i+8], setj[i+8]]
      arr[seti[i], setj[i]]=arr[seti[i], setj[i]]-arr[seti[i+8], setj[i+8]]

    n=n+1

    ones.append(0)

    for i in range(120):
      for j in range(125):
        if(arr[i][j]==1):
          ones[n]=ones[n]+1

    x.append(n)
    y.append(ones[n]-ones[n-1])
     
    if(ones[n]==15000):
      break

  plt.plot(x, y)
  yhat = savgol_filter(y, 5, 1)
  plt.plot(x, yhat, color = 'red')
  plt.xlabel("#iteration")
  plt.ylabel("dy/dx")
  plt.savefig("SoC-assgn2-Plot-3-Prasann.jpg")
  print(max(y))
```
